```python
import pandas as pd
```

```python
#1. Obtain data
data = pd.read_csv("your_own_path_here/facebook checkin/train.csv")
```

```python
data
```

```python
#2. Basic data preprocessing
# 1) Narrow down the range of data
data = data.query("x<0.4 & x>0 & y<0.4 & y>0 ")
```

```python
data
```

```python
data.head()
```

```python
# 2) Time feature processing
time_value = pd.to_datetime(data["time"], unit = "s")
```

```python
time_value.values
```

```python
date = pd.DatetimeIndex(time_value)
date
```

```python
data["day"] = date.day
```

```python
data["weekday"] = date.weekday
```

```python
data["hour"] = date.hour
```

```python
data
```

```python
# 3) Filter locations that are not visited often
place_count = data.groupby("place_id").count()["row_id"]
```

```python
place_count[place_count > 3]
```

```python
data_final = data[data["place_id"].isin(place_count[place_count > 3].index.values)]
```

```python
# Filter features and targets
x = data_final[["x", "y", "accuracy", "day", "weekday", "hour"]]
y = data_final["place_id"]
```

```python
x.head()
```

```python
y.head()
```

```python
# Data split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
    #3 Preprocessing：standardization
    transfer = StandardScaler()
    x_train = transfer.fit_transform(x_train)
    x_test = transfer.transform(x_test)
    
    #4 KNN estimator
    estimator = KNeighborsClassifier(n_neighbors = 4)
    
    #grid search and cross validation 
    #prepare parameters
    param_dict = {"n_neighbors": [3, 5, 7, 9]}
    estimator = GridSearchCV(estimator, param_grid = param_dict, cv = 5)
    estimator.fit(x_train, y_train)
    
    #5 model assessment
    #method 1：compare real and predicted values
    y_predict = estimator.predict(x_test)
    print("y_predict: \n", y_predict)
    print("Comparing real and predicted values: \n", y_test == y_predict)
    
    #method 2：compare accuracy
    score = estimator.score(x_test, y_test)
    print("Accuracy is： \n", score)
    
    # Best parameter：best_params_
    print("Best parameter：\n", estimator.best_params_)
    # Best score：best_score_
    print("Best score：\n", estimator.best_score_)
    # Best estimator：best_estimator_
    print("Best estimator：\n", estimator.best_estimator_)
    # Cross validation：cv_results_
    # print("Cross validation：\n", estimator.cv_results_)
```
